```
def fib(n):
    a=0
    b=1
    if n==1:
        print(a)
    elif n<1:
        print('not found')
    else:
        print(a)
        print(b)
        for i in range(2,n):
            c=a+b
            a=b
            b=c
            print(c)

fib(15)
```

```
def even():

    for i in range(1999,3001):
        if i%2==0:
            print(i)


even()
```

```
def mul(n):
    for i in range(1,n):
        for j in range(1,n):
            print(i*j, end="\t")
        print("\n")


mul(12)
```

```
def num(n):

    if n%2==0:
        print('even')
    else:
        print('odd')



num(88)


```

```
def palindrome():

    a = input('enter the number')
    b = a[::-1]

    if a==b:
        print('the number is palindrome')
    else:
        print('not a palindrome')

palindrome()
```

```
def prime(n):

    for i in range(2,n):
        if n%i==0:
            print('its not a prime')
            break
    else:
        print('its a prime')

prime(13)
```

```
def square(n):

    for i in range(n):
        a = i**2
        print(a)

square(5)
```
